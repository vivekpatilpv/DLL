```python
# Importing the libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
```

```python
# Reading the dataset
data = pd.read_csv("/content/diabetes1_assign3.csv")
```

```python
# Splitting into X part of diabetes data i.e. input part
x = data.iloc[:,:8]
```

```python
y = data["Outcome"]
```

```python
std = StandardScaler()
x_std = std.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_std,y,test_size = 0.1)
```

Creating a model for optimizer as Adam

```python
adam_model=models.Sequential()
# adam_model.add(layers.Dense(100,activation="relu"))      # First hidden layer
adam_model.add(layers.Dense(100,activation="relu")) 
# adam_model.add(layers.Dense(50,activation="relu"))       # second hidden layer
adam_model.add(layers.Dense(50,activation="relu")) 
# adam_model.add(layers.Dense(12,activation="relu"))       # third hidden layer
adam_model.add(layers.Dense(25,activation="relu")) 
# adam_model.add(layers.Dense(8,activation="relu"))        # fourth hidden layer
adam_model.add(layers.Dense(12,activation="relu"))
adam_model.add(layers.Dense(6,activation="relu"))  
adam_model.add(layers.Dense(1,activation="sigmoid"))     # output layer
```

```python
adam_model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history=adam_model.fit(x_train,y_train,epochs=50, validation_data=(x_test,y_test))
```

```python
result=adam_model.evaluate(x_test,y_test)
print(f"Loss in the model :-  {result[0]}")
print(f"Accuracy of the model :- {result[1]*100}")
```

```python
# Graph of how loss is decreasing with the epochs
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.ylim([0, 0.8])
plt.legend(loc='lower right')
test_loss, test_acc = adam_model.evaluate(x_test, y_test, verbose=2)
plt.plot(history.history['loss'], label='loss')
```

Creating model for optimizer as RMSprop

```python
RMSprop_model=models.Sequential()
RMSprop_model.add(layers.Dense(100,activation="relu"))      # First hidden layer
RMSprop_model.add(layers.Dense(50,activation="relu"))       # second hidden layer
RMSprop_model.add(layers.Dense(12,activation="relu"))       # third hidden layer
RMSprop_model.add(layers.Dense(8,activation="relu"))        # fourth hidden layer
RMSprop_model.add(layers.Dense(1,activation="sigmoid"))     # output layer
```

```python
RMSprop_model.compile(optimizer="RMSprop",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history=RMSprop_model.fit(x_train,y_train,epochs=50, validation_data=(x_test,y_test))
```

```python
result=RMSprop_model.evaluate(x_test,y_test)
print(f"Loss in the model :-  {result[0]*100}")
print(f"Accuracy of the model :- {result[1]*100}")
```

```python
# Graph of how loss is decreasing with the epochs
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.ylim([0.3, 0.8])
plt.legend(loc='lower right')
test_loss, test_acc = RMSprop_model.evaluate(x_test, y_test, verbose=2)
plt.plot(history.history['loss'], label='loss')
```
